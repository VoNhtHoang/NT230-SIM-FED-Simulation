````python

#################### import ##########################
import hashlib
import dask
from dask_ml.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from dask.dataframe import isna
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import Pipeline
import os
#################### import ##########################

####### func ##########
def ip_to_float(ip):
    try:
        return float(int(ipaddress.IPv4Address(ip)))/1e9
    except:
        return 0.0  # for invalid or empty IPs
    
def sum_of_squares(partition):
    return pd.Series([(partition ** 2).sum()])

def string_to_float(s):
    if pd.notna(s):
        return int(hashlib.sha256(s.encode('utf-8')).hexdigest(), 16) % 10**8 / 1e8
    return 0
####### func #########

cols_to_float = ['duration']
cols_to_int = [ 'orig_bytes', 'resp_bytes', 'missed_bytes', 'orig_pkts', 'orig_ip_bytes', 'resp_pkts', 'resp_ip_bytes']

df = df.drop(columns = ['uid', 'ts', 'tunnel_parents', 'local_orig', 'local_resp'])
print("Done drop timestamp, uid")

df = df.replace(["(empty)", "-"], np.nan)
df = df.drop_duplicates()
print("Done dropdup")
````

````python
label_mapping = {"benign": 0, "Malicious": 1, "Benign": 0, "malicious": 1}
df["label"] = df["label"].map(label_mapping).fillna(-1).astype("int32")

df["id.orig_h"] = df["id.orig_h"].apply(lambda x: ip_to_float(x)).astype("float32")
df["id.resp_h"] = df["id.resp_h"].apply(lambda x: ip_to_float(x)).astype("float32")

df['conn_state'] = df['conn_state'].apply(lambda x: string_to_float(x)).astype("float32")
df['id.resp_p'] = df['id.resp_p'].apply(lambda x: string_to_float(x)).astype("float32")
df['id.orig_p'] = df['id.orig_p'].apply(lambda x: string_to_float(x)).astype("float32")
df['history'] = df['history'].apply(lambda x: string_to_float(x)).astype("float32")

# df['conn_state'] = df['conn_state'].apply(lambda x: string_to_float(x))
# df['id.resp_p'] = df['id.resp_p'].apply(lambda x: string_to_float(x))
# df['id.orig_p'] = df['id.orig_p'].apply(lambda x: string_to_float(x))
# df['history'] = df['history'].apply(lambda x: string_to_float(x))
print("Done encode string to float")

df = dd.get_dummies(
    df,
    columns=['proto', 'service'],
    prefix={'proto': 'proto', 'service':'service'},  # Custom prefix
    # dummy_na=True,  # Tạo cột cho giá trị NaN
    # drop_first=True,  # Bỏ đi một cột để tránh đa cộng tuyến
    dtype="int32"
)
print("Done onehot")

df=df.drop(columns = ['service_-'])

df = df.fillna("0")
for col in cols_to_float:
    df[col] = df[col].astype('float32')
for col in cols_to_int:
    df[col] = df[col].astype('float32')
````

````python
def normalize_partition(pdf):
    # Tính chuẩn L2 cho từng hàng trong partition
    row_norm = np.sqrt((pdf**2).sum(axis=1))
    # Tránh chia cho 0
    row_norm[row_norm == 0] = np.nan
    # Chuẩn hóa từng hàng theo L2 norm
    return pdf.div(row_norm, axis=0)

df[numeric_cols] = df[numeric_cols].map_partitions(normalize_partition, meta=dict(df[numeric_cols].dtypes))
print("Done Normal")
````
